# Newton's Laws of Gravitation

The force of gravity experienced by two bodies $i$ and $j$ can be expressed by the following equation

$\mathbf{F} = \frac{Gm_i m_j}{r_{ij}^3} \mathbf{r}_{ij}$

Extending this to Newton's Second Law $\sum \mathbf{F}_i = m_i \mathbf{a}_i $, the Resultant motion due to a force purely governed by this gravitational attraction can be found following...

$\mathbf{F}_i = \frac{Gm_i m_j}{r_{ij}^3} (\mathbf{r}_{j} - \mathbf{r}_{i})$
$\implies m_i \mathbf{a}_i = \frac{Gm_i m_j}{r_{ij}^3} (\mathbf{r}_{j} - \mathbf{r}_{i})$

The direction of vector $\mathbf{r}_{ij}$ depends on the body. We know gravity is an attractive force, hence must be directed towards the position of the opposing body. By reducing the expression, noticing the independence of the body in question's mass, we find ...

$ \ddot{\mathbf{r}}_i = \frac{G m_j}{r_{ij}^3} (\mathbf{r}_{j} - \mathbf{r}_{i})$, and similarly
$ \ddot{\mathbf{r}}_j = \frac{G m_i}{r_{ij}^3} (\mathbf{r}_{i} - \mathbf{r}_{j})$

These equations are perfectly valid for determining the resultant motion of each body, but their coupled nature makes it difficult to solve analytically. A change in perspective can help us simplify the equation further. Rather than solving for the exact motion of the bodies in a fixed frame, it is often more convienent to solve for the motion of one body relative to another. In this perspective, we can define relative positions, velocities and accelerations as the difference between their individual vectors.

i.e. $\mathbf{r}_{ij} = \mathbf{r}_{j} - \mathbf{r}_{i},$
$\quad \dot{\mathbf{r}}_{ij} = \dot{\mathbf{r}}_{j} - \dot{\mathbf{r}}_{i},$
$\quad \ddot{\mathbf{r}}_{ij} = \ddot{\mathbf{r}}_{j} - \ddot{\mathbf{r}}_{i}.$ 

Hence, solving for $\ddot{\mathbf{r}}_{ij}$...

$$\ddot{\mathbf{r}}_{ij} = \frac{G m_i}{r_{ij}^3} (\mathbf{r}_{i} - \mathbf{r}_{j}) 
 - \frac{G m_j}{r_{ij}^3} (\mathbf{r}_{j} - \mathbf{r}_{i})$$

$$\begin{align*}
\ddot{\mathbf{r}}_{ij}  &= \frac{G m_i}{r_{ij}^3} (\mathbf{r}_{i} - \mathbf{r}_{j}) 
                            - \frac{G m_j}{r_{ij}^3} (\mathbf{r}_{j} - \mathbf{r}_{i}) \\
                        &= \frac{G m_i}{r_{ij}^3} (- \mathbf{r}_{ij}) 
                            - \frac{G m_j}{r_{ij}^3} (\mathbf{r}_{ij}) \\
                        &= -\left( \frac{G (m_i + m_j)}{r_{ij}^3} \right) \mathbf{r}_{ij} \\
\end{align*}$$

We can define a gravitational attraction parameter $\mu = G(m_i + m_j)$, to retrieve the common solution form...

$$ \ddot{\mathbf{r}}_{ij} + \frac{\mu}{r_{ij}^3} \mathbf{r}_{ij} = \mathbf{0}$$




# Manipulating Our Equation

While this equation describes the acceleration of a body w.r.t. another at any given displacement, it can still be difficult to predict the overall motion expected between them. We can manipulate the equation to find some core properties, and hopefully, find a method to quickly determine the overall trajectory of our bodies.

We start by taking a L-hand cross product with the bodies relative displacement vector...

$$ \mathbf{r}_{ij} \times \ddot{\mathbf{r}}_{ij} + \frac{\mu}{r_{ij}^3} \mathbf{r}_{ij} \times \mathbf{r}_{ij} = \mathbf{0} $$

Consider the the derrivative of $\mathbf{r}_{ij} \times \dot{\mathbf{r}}_{ij}...$

$$
\frac{\partial}{\partial t} (\mathbf{r}_{ij} \times \dot{\mathbf{r}}_{ij}) = \dot{\mathbf{r}}_{ij} \times \dot{\mathbf{r}}_{ij} + \mathbf{r}_{ij} \times \ddot{\mathbf{r}}_{ij}
$$

Recalling the property $\mathbf{a} \times \mathbf{a} = \mathbf{0},$ we can simplify both expressions and substitute to find...

$$
\frac{\partial}{\partial t} (\mathbf{r}_{ij} \times \dot{\mathbf{r}}_{ij}) = \mathbf{0}.
$$

Hence, integrating provides us a useful result...

$$
\begin{align*}
\mathbf{r}_{ij} \times \dot{\mathbf{r}}_{ij} &= \boldsymbol{\kappa} \\
                                             &= \mathbf{h}.
\end{align*}
$$

We find our arbitary integration constant ($\boldsymbol{\kappa}$), is the specific angular momentum of the orbit which we provide a special constant label to as $\mathbf{h}$. This must remain constant at all points of the orbit, provided a 2-body problem driven by purely gravitational forces.

Now, lets take the R-hand cross product of our gravitational equation and our new angular momentum vector...
$$
\begin{align*}
\ddot{\mathbf{r}}_{ij} \times \mathbf{h} + \frac{\mu}{r_{ij}^3} \mathbf{r}_{ij} \times \mathbf{h} = \mathbf{0}
\end{align*}
$$

Consider the derrivative of $\dot{\mathbf{r}}_{ij} \times \mathbf{h}...$

$$
\begin{align*}
\frac{\partial}{\partial t} (\dot{\mathbf{r}}_{ij} \times \mathbf{h}) &= \ddot{\mathbf{r}}_{ij} \times \mathbf{h} + \sout{\dot{\mathbf{r}}_{ij} \times \dot{\mathbf{h}}} \\
    &= \ddot{\mathbf{r}}_{ij} \times \mathbf{h}
\end{align*}
$$

Now we can inspect the second term closer. Lets start by using the remembering that $\mathbf{h} = \mathbf{r}_{ij} \times \dot{\mathbf{r}}_{ij},$ then applying the vector identity $ABC = BAC - CAB$ rule...

$$
\begin{align*}
\frac{\mu}{r_{ij}^3}\mathbf{r}_{ij} \times \mathbf{h} 
    &= \frac{\mu}{r_{ij}^3} \mathbf{r}_{ij} \times \mathbf{r}_{ij} \times \dot{\mathbf{r}}_{ij} \\
    &= \frac{\mu}{r_{ij}^3} \left[\mathbf{r}_{ij}(\mathbf{r}_{ij} \cdot \dot{\mathbf{r}}_{ij}) - \dot{\mathbf{r}}_{ij}(\mathbf{r}_{ij} \cdot \mathbf{r}_{ij})\right] \\
    &= \frac{\mu}{r_{ij}^3} [\mathbf{r}_{ij}(r_{ij}v_{r}) - \dot{\mathbf{r}}_{ij}(r_{ij}^2)] \\
\end{align*}
$$

By simplifying from the dot product (where $v_r$ is the radial component of velocity, hence also implies $v_r = \dot{r}_{ij}$), and further considering the definition of the quotient rule ($\frac{\partial}{\partial t}(\frac{u}{v}) = \frac{\dot{u}v - u\dot{v}}{v^2}$), we find...

$$
\begin{align*}
\frac{\mu}{r_{ij}^3}\mathbf{r}_{ij} \times \mathbf{h} 
    &= \mu \left[\frac{\mathbf{r}_{ij}\dot{r}_{ij} - \dot{\mathbf{r}}_{ij}r_{ij}}{r_{ij}^2} \right] \\
    &= - \mu \left[\frac{\dot{\mathbf{r}}_{ij}r_{ij} - \mathbf{r}_{ij}\dot{r}_{ij}}{r_{ij}^2} \right] \\
    &= -\mu \left[\frac{\partial}{\partial t}\left(\frac{\mathbf{r}_{ij}}{r_{ij}} \right) \right]
\end{align*}
$$

Hence, we can integrate the full equation to find...

$$
\begin{align*}
\mathbf{0} 
    &= \ddot{\mathbf{r}}_{ij} \times \mathbf{h} + \frac{\mu}{r_{ij}^3} \mathbf{r}_{ij} \times \mathbf{h} \\
    &= \frac{\partial}{\partial t}\left(\dot{\mathbf{r}}_{ij} \times \mathbf{h} - \mu \frac{\mathbf{r}_{ij}}{r_{ij}} \right) \\
\boldsymbol{\kappa}_2
    &= \dot{\mathbf{r}}_{ij} \times \mathbf{h} - \mu \frac{\mathbf{r}_{ij}}{r_{ij}} \\
    &= \mathbf{\mu e}
\end{align*}
$$

This new arbitary integration constant ($\boldsymbol{\kappa}_2$), is the eccentricity vector of the orbit which we provide the constant label $\mathbf{e} = \frac{\mathbf{v} \times \mathbf{h}}{\mu} - \frac{\mathbf{r}}{r}$.

So far, we have defined multiple vector quantites. While accurate, they are still slightly laboratrus for analysing the overall geometry of our orbit. It could be easier to understand the properties of an orbit, by considering the evolution of an overall scalar quantity (i.e. radial distance instead of displacement). We can do this by taking the dot product $\mathbf{r} \cdot \mathbf{e}$...

$$
\begin{align*}
\mathbf{r} \cdot \mathbf{e} &= \frac{\mathbf{r}\cdot(\mathbf{v} \times \mathbf{h})}{\mu} - \frac{\mathbf{r} \cdot \mathbf{r}}{r} \\
re\cos{\theta} &= \frac{\mathbf{r}\cdot(vh\sin{(90)} \: \hat{\mathbf{n}})}{\mu} - \frac{r^2}{r} \\
    &= \frac{rvh\cos{\psi}}{\mu} - r \\
    &= \frac{rv_T h}{\mu} - r \\
    &= \frac{h^2}{\mu} - r \\
\implies r(1 + e\cos{\theta}) &= \frac{h^2}{\mu} \\
\therefore r &= \frac{h^2}{\mu}\frac{1}{1 + e\cos{\theta}}
\end{align*}
$$

This derivation was quite involved and used many steps. We started by using the dot product rule, and cross product rule. By the definition of our angular momemntum vector ($\mathbf{h} = \mathbf{r} \times \mathbf{v}$), we should expect the angle between $\mathbf{h}$ and $\mathbf{v}$ to be $90\deg$. The resultant normal vector from this cross product will be orthogonal to both angular momentum and velocity. If our displacement is not already orthogonal to velocity, then there will be an angle between it and the normal vector. This is called the Flight Path Angle (FPA or $\psi$) and the cosine of this angle determines the orthogonal component of displacement or velocity to the other.
Using this we can then compute the tangential velocity, and by definition reduce the multiplication of the tangential velocity and radial distance to the angular momentum magnitude. From here it is simple re-arranging to isolate $r$ in terms of the eccentricity $e$ and an angle called the true anomoly $\theta$, the angle between the radial vector and the eccentricity vector.
